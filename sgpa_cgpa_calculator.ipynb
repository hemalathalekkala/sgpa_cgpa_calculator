#####     SGPA_CGPA_CALCULATOR

```python
sem=int(input("Enter Number of Semesters:"))

cgpa=0

for i in range(1,sem+1):
    
    sgpa=0
    
    total_credits=0
    
    sub=int(input(f"Enter Number of Subjects for Semester {i}:"))
    
    for j in range(1,sub+1):
        
        points=int(input("Enter Scored Points:"))
        
        credits=int(input("Enter Credits:"))
        
        sgpa=sgpa+(points*credits)
        
        total_credits=total_credits+credits
        
    print(f"Total Credits for {i} Semester:",total_credits)
        
    sgpa=sgpa//total_credits
    
    print(f"Semester {i} SGPA:",round(sgpa,2))
    
    cgpa=cgpa+sgpa
    
cgpa=cgpa//sem

print("Final CGPA:",cgpa)
if cgpa>=9.0:
    print("Outstanding")
    print("Grade: A")
elif cgpa>=8.0:
    print("Excellent")
    print("Grade: B")
elif cgpa>=7.0:
    print("Very Good")
    print("Grade: C")
elif cgpa>6.0:
    print("Good") 
    print("Grade: D")
else:
    print("Needs Improvement")
    print("Grade: F")
```
